```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (Replace 'your_dataset.csv' with the actual filename)
df = pd.read_csv('listing_data_publish.csv')

# Display first few rows of the dataset
print("\nðŸ”¹ First 5 Rows of Dataset:")
print(df.head())

# Check dataset shape (rows, columns)
print(f"\nðŸ”¹ Dataset Shape: {df.shape[0]} rows, {df.shape[1]} columns")

# Check column names and data types
print("\nðŸ”¹ Dataset Info:")
print(df.info())

# Check for missing values in each column
print("\nðŸ”¹ Missing Values in Each Column:")
print(df.isnull().sum())

# Summary statistics for numerical features
print("\nðŸ”¹ Summary Statistics (Numerical Features):")
print(df.describe())

# Summary statistics for categorical features
print("\nðŸ”¹ Summary Statistics (Categorical Features):")
print(df.describe(include=['object']))

# Identify categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns.tolist()
print(f"\nðŸ”¹ Categorical Columns: {categorical_columns}")

# Identify numerical columns
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
print(f"\nðŸ”¹ Numerical Columns: {numerical_columns}")

# Analyze the distribution of listings across different categories (if 'Category' exists)
if 'Category' in df.columns:
    plt.figure(figsize=(10, 5))
    sns.countplot(y=df['Category'], order=df['Category'].value_counts().index, palette="viridis")
    plt.title('Distribution of Listings Across Categories')
    plt.xlabel('Count')
    plt.ylabel('Category')
    plt.show()

# Analyze the distribution across different regions (if 'Region' exists)
if 'Region' in df.columns:
    plt.figure(figsize=(10, 5))
    sns.countplot(y=df['Region'], order=df['Region'].value_counts().index, palette="coolwarm")
    plt.title('Distribution of Listings Across Regions')
    plt.xlabel('Count')
    plt.ylabel('Region')
    plt.show()

# Display unique values in categorical features
print("\nðŸ”¹ Unique Values in Categorical Columns:")
for col in categorical_columns:
    print(f"{col}: {df[col].nunique()} unique values")

```
